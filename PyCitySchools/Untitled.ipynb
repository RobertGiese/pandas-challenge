```python
import pandas as pd
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_count = len(school_data_complete["school_name"].unique())
student_count = school_data_complete["Student ID"].count()
total_budget = school_data_complete["budget"].sum()
avg_math_score = school_data_complete['math_score'].mean()
avg_reading_score = school_data_complete['reading_score'].mean()
passing_math = school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]
pct_passing_math = passing_math/student_count*100
passing_reading = school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
pct_passing_reading = passing_reading/student_count*100
overall_passing_rate = (pct_passing_math + pct_passing_reading)/2
avg_math_score = round(avg_math_score, 2)
avg_reading_score = round(avg_reading_score, 2)
pct_passing_math = round(pct_passing_math, 2)
pct_passing_reading = round(pct_passing_reading, 2)
overall_passing_rate = round(overall_passing_rate, 2)
district_summary = pd.DataFrame({"Total Schools" : [school_count],
                                 "Total Students" : [student_count],
                                 "Total Budget" : [total_budget],
                                 "Average Math Score" : [avg_math_score],
                                 "Average Reading Score" : [avg_reading_score],
                                 "% Passing Math" : [pct_passing_math],
                                 "% Passing Reading" : [pct_passing_reading],
                                 "% Overall Passing Rate" : [overall_passing_rate]})
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary
```

```python
school_name = school_data ["school_name"]
school_type = school_data ["type"]
student_count = school_data ["size"]
total_school_budget = school_data ["budget"]
per_student_budget = total_school_budget/student_count
per_student_budget = per_student_budget.map("${:,.2f}".format)
school_summary = pd.DataFrame({"School Name" : school_name,
                               "SchoolType" : school_type,
                               "Total Students" : student_count,
                               "Total School Budget" : total_school_budget,
                               "Per Student Budget" : per_student_budget})
school_summary.set_index("School Name")
school_summary
```
